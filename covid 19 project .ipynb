```python
df['Recovered Rate']=(df['Recovered']/df['Confirmed'])*100
df['Recovered Rate'].replace(np.inf,0,inplace=True)#numerical value infinite ahh vandha '0' va convert pannanum
df['Recovered Rate'].fillna(0,inplace=True)#missing values ahh '0' va replace pannanum

highest_recovery_rate=df.loc[df['Recovered Rate'].idxmax()]
print('Country with highest_recovery_rate :',highest_recovery_rate['Country/Region'])
```

```python
I want to get the Highest deathing rate country:
```

```python
df['Mortality Rate']=(df['Deaths']/df['Confirmed'])*100
df['Mortality Rate'].replace(np.inf,0,inplace=True)#numerical value infinite ahh vandha '0' va convert pannanum
df['Mortality Rate'].fillna(0,inplace=True)#missing values ahh '0' va replace pannanum

highest_Mortality_Rate=df.loc[df['Mortality Rate'].idxmax()]
print('Country with highest_Mortality_Rate :',highest_Mortality_Rate['Country/Region'])
```

```python
import dash
from dash import dcc,html
import plotly.express as px
import plotly.graph_objects as go
print('dash is installed and imported')
```

```python
chart1 =px.bar(df.groupby('WHO Region')['Confirmed'].sum().reset_index(),
               x='WHO Region',y='Confirmed',title='Total Confirmed Cases by WHO Region')

chart2 =px.bar(df.sort_values('Active',ascending=False).head(10),
               x='Country/Region',y='Active',title='Top 10 Active Countries')

chart3 =px.pie(df,values='Confirmed',names='WHO Region',title='Confirmed Cases by WHO Region')

chart4 =px.scatter(df,x='Confirmed',y='Deaths',color='WHO Region',
               title='Confirmed Cases vs Deaths')

chart5 =px.histogram(df,x='Recovered Rate',nbins=20,title='Distribution of Recovery Rates across counties')

chart6 =px.bar(df,x='WHO Region',y='Mortality Rate',title='Mortality Rate by WHO Region')

correlation=df[['Confirmed','Deaths','Recovered','Active']].corr()
chart7 = go.Figure(data=go.Heatmap(x=correlation.columns,
                                  y=correlation.columns,
                                  z=correlation.values,
                                   colorscale='viridis'))

chart7.update_layout(title='Correlation Heatmap')

chart8 =px.box(df,x='WHO Region',y='Deaths / 100 Cases',title='Deaths per 100 Cases by WHO Region')

chart9 =px.scatter(df,x='WHO Region',y='1 week change',color='WHO Region',title='Weekly COVID Case Change vs Last Week')

chart10 =px.violin(df,x='WHO Region',y='Deaths / 100 Cases',title='Modality Rate Distribution by WHO Region')
```

```python
app =dash.Dash(__name__)

app.layout = html.Div([
    html.H1("COVID 19 DATA DASHBOARD", style={'textAlign': "center"}),

    dcc.Graph(figure=chart1),
    dcc.Graph(figure=chart2),
    dcc.Graph(figure=chart3),
    dcc.Graph(figure=chart4),
    dcc.Graph(figure=chart5),
    dcc.Graph(figure=chart6),
    dcc.Graph(figure=chart7),
    dcc.Graph(figure=chart8),
    dcc.Graph(figure=chart9),
    dcc.Graph(figure=chart10),

    html.P("Build by k7", style={'textAlign': 'center', 'marginTop': '20px'})
 ])
```

```python
if __name__=='__main__':
    app.run(debug=True)
```
